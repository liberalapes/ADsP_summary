# 과목 IV 데이터 분석

## 제 2장 통계 분석

### 제4절 시계열 예측

#### 1. 정상성

- 시계열 자료(Time-series Data) : 시간의 흐름에 따라서 관측된 데이터
- 장싱상(stationary) : 시계열 분석(Time-series Analysis)를 위해서 만족해야 하는 것
    - 시점에 상관없이 시계열의 특성이 일정하다는 것을 의미

- 정상성
    - 평균이 일정하다
    - 분산이 시점에 의존하지 않는다
    - 공분산은 단지 시차에만 의존하고 시점자체에 의존하지 않는다

- 시계열 자료를 비정상 시계열 : 정의한 정상정 조건을 하나라도 만족하지 못하는 경우
    - 대부분의 시계열 자료는 비정상 시계열 자료
    - 비정상 시계열 자료는 정상성을 만족하도록 데이터를 정상 시계열 자료로 만든 후에 식열 분석을 수행
- 이를 위해서 정상성을 만족하는지 검증이 필요
- 시계열 자료 그림을 통해 자료의 이상점(Outlier)과 개입(Intervention)을 살피고, 정상성 만족 여부와 개략적인 추세 유무를 관찰
- 추세를 보이는 (평균이 일정하지 않은) 경우에는 차분(Difference)을 통해 비정상 시계열을 정상 식열로 바꿈
- 시간예 따라 분선이 일정하지 않은 경우에는 변환(Trancsformation)을 통해 정상 시계열로 변경

- 차분 : 현 시점의 자료값에서 전 시점의 자료값을 빼는 것
- 일반적인 차분은 현재 시점에서 바로 전 시점의 자료값을 빼는 것을 말하며, 여러 시점 전의 자료를 빼는 것을 계절차분(Seasonal Diffrence)이라고 함
- 계절성을 갖는 비정상 시계열을 정상 시계열로 바꿀 때 계절차분을 사용

#### 2. 시계열 모형

##### 가. 자기회귀 모형(AR 모형)

- 자기 회귀 모형(Autoregressive model) : 현 시점의 자료가 p 시점 전의 과거 자료로 설명될 수 있다는 의미
    - AR(p) 모형

$$ Z_t = \phi Z_{t-1} + \phi _2 Z_{t-2} + \phi _p Z_{[-1} + \alpha _t$$

Z_t :  현재 시점의 시계열 자료 <br>
Z_{t-1}, Z_{t-2} : 1~p 시점 이전의 시계열 자료<br>
φ_p : p 시점이 현재 시점에 어느 정도 영향을 주는지 나타내는 모수<br>
a_t : 백색잡음과정(White noise process, 대표적 정상 시계열), 시계열 분석에서 오차항을 의미

- 백색잡음과정 : 대표적 정상 시계열
    - 백색잡음과정 a_t는 독립이고 같은 분포를 따르며, 평균이 0이고 분산이 a^2_a인 확률변수
    $$ a_t = ~ iid(0, a^2_a)$$

- 자기회귀모형 : 현 시점의 시계열 자료에 몇 번쨰 전 자료까지 영향을 주는지 알아냄
- 현 시점의 시계열 자료에 과거 1 시점 이전의 자료만 영향을 준다면, 이를 1차 자기회귀모형이라고 하며, AR(1) 모형이라 함
$$ Z_t = \phi Z_{t-1} + a_t$$

- 현 시점의 시계열 자료에 과거 2시점 전 자료까지 영향을 주는 모형을 2차 자기회기모형이라고 함, AR(2) 모형은 다음과 같음
$$ Z_t = \phi Z_{t-1} + a_t $$
- 자기회귀 모형인지 판단하기 위한 모형 식별을 위해서, 자료에서
    - 자기상관함수(ACF, Auto-Correlation Function)과
    - 부분자기상관함수(PACF, Partial Auto-Correlation Function)을 이용하여 식별한다.
- 일반적으로
    - 자기회귀모형은 자기상관함수는 시차가 증가함에 따라 점차적으로 감소하고,
    - 부분자기상관함수는 p+1 시차 이후 급격히 감소하여 절단된 형태이며, 이를 AR(p) 모형이라 판별함.

##### 나. 이동평균모형(AM 모형)

- 시계열 자료를 모형화 하는데 자기회귀모형 다음으로 많이 쓰이는 모형이 이동평균모형(Moving Average model)이다.
$$ Z_t = a_t - \theta _1{a_{t-1}} - \theta _2{a_{t-2}} - ... - \theta _p{a_{t-p}}$$
- 이동평균모형은 현 시점의 자료를 유한 개의 백색잠음의 선형결합으로 표현되었기 때문에 항상 정상서을 만족함
    - 때문에 이동평균모형은 정상성 가정이 필요없음
- 1차 이동평균모형,(MA(1) 모형은 가장 간단한 이동평균모형으로 같은 시점의 백색잡음과 바로 전 시점의 백색잡음의 결합으로 이루어진 모형이다.
    $$ Z_t = a_t - \theta _1 a_{t-1}$$
- 다음으로 2차 이동평균모형인 MA(2) 모형은 아래와 같음.
    $$ Z_t = a_t - \theta _1 a_{t-1} - \theta _2 a_{t-2}$$

- 이동평균모형을 판단하기 위한 모형 식별은자기회귀모혀과 마찬가지로 자기상관함수와 부분자기상관함수를 이용하여 식별하게 됨
- 이동평균모형은 자기회귀모형과 반대로 자기상관함수는 p+1 시차 이후 절단된 형태가 되고, 이때를 MA(p) 모형이라고 본다.
- 그리고 부분자기상관함수는 점차 감소하는 형태를 띄게 됨

##### 다. 자기회귀누적 이용동평균모형(ARIMA 모형)

- 대부분의 많은 시계열 자료가 자기회귀누적이동평균모형(Autoregressice Intergrated Movinf Average Model)을 따름
- ARIMA 모형은 기본적으로 비정상 시계열 모형이기 때문에 차분이나 변환을 통해해 AR 모형이나 MA 모형, ARMA 모형으로 정상화 할 수 있음
- ARIMA(p,d,q) 모형은 차수 p,d,q의 값에 따라 모형의 이름이 다르게 됨
- 차수 p는 AR 모형과 관련이 있고, q는 MA모형과 관련이 있는 차수
- d는 MA 모형과 관련있는 차수
- d는 ARIMA 에서 ARMA로 정상화할 때 몇번 차분을 했는지를 의미함
- d = 0이면 ARMA(p,q) 모형이라고 부르고, 이 모형은 정상성을 만족함
- p = 0 이면 IMA(d,g) 모형이라고 부르고, 이 모형은 d번 차분한 시계열 모형이 AR(p) 모형을 따르게 됨    

##### 라. 분해 시계열

- 분해 시계열 : 시계열에 영향을 주는 일반적인 요인을 시계얄에서 분리해 분석하는 방법
    - 회귀분석적인 방법을 주로 사용
- 시계열을 구성하는 요소는 다음 4가지로 분류

###### 1) 추세요인

- 자료의 그림을 그렸을 때, 그 형태가 오르거나 내리는 추세를 따르는 경우
- 선형적으로 추세가 있는 것 이외에도 이차식의 형태를 취하거나 지수적 형태를 취할 수 있음.
- 이렇게 자료가 어떤 특정한 형태를 취할 때 추세요인(Trend factor)이 있다고 함

###### 2) 계절요인

- 요일마다 반복되거나 일년 중 각 월에 의한 변화, 사분기 자료에서 각 분기에 의한 변화 등 고정된 주기에 따라 자료가 변화하는 경우
- 고정된 주기에 따라 자료가 변화할 경우, 계절요인(Seasonal factor)가 있다고 함

###### 3) 순환요인

- 명백한 경제적이나 자연적인 이유가 없이 알려지지 않은 주기를 가지고 변화하는 자료
- 알려지지 않은 주기를 가지고 자료가 변화할 때, 순환요인(Cyclical factor)가 있다고 함

###### 4) 불규칙요인

- 위 세가지 요인으로 설명 할 수 없는 회귀분석에서 오차에 해당하는 요인을 불규칙요인(Inrregular factor)라고 함

- 분해 시계열 분석법에서는 각 구성요인을 정확하게 분리하는 것이 중요
- 그러나 각 요인을 정확하게 분리하는 것은 쉽지 않음
- 또한 분해 시계열 방법은 이론적인 약점이 있는 것도 있음
- 하지만 경제학자들이나 조사통계학자들은 이러한 약점에도 불구하고 널리 사용 중
- 실제로 경제 분석이나 예측에서 이 방법은 성공적으로 사용되는 중
- 분해식의 일반적 정의는 다음과 같음
$$ Z_t = f(T_t,S_t,C_t,I_t$$
