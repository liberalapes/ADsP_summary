#### 1. 상관 분석

- 상관분석(Correlation Analysis) : 데이터 안의 두 변수 간의 관계를 알아보는 분석
- 상관계수(Correlation coefficient) 
    - 피어슨 상관계수(Pearson correlation) : 등간척도 이상으로 측정되는 두 변수들 간의 상관관계를 측정
    - 스피어만 상관계수(Spearman's rank correlation coefficient) : 서열척도인 두 변수들의 상관관계를 측정

- 상관계수가 양수 : x가 증가할 때, y도 증가
- 상관계수가 음수 : x가 증가할 때, y는 감소
- 산점도가 직선에 가깝게 분포하면, 상관계수의 절대값이 1에 가까워짐 : 강한 상관 관계
- 산점도가 직선에서 넓게 펴져 x와 y의 직선상의 관계가 희미할 수록 상관계수는 0에 가까워짐 : x의 변동에 따라 y의 패턴에 변화가 없음

##### 가. 피어슨의 상관계수

- 두 확률변수 X와 Y의 선형관계의 크기를 측정하는 공분산은 아래와 같다.
$$ Cou(X,Y) = E[(X-\mu_x)(Y-\mu_y)] $$
- μx 와 μy는 X, Y의 기댓값
- 공분산의 크기 : X와 Y의 단위에 따라 영향을 받는데, 단위에 영향을 받지 않고 -1과 1사이의 값을 갖도록 정의 
$$ \rho = Corr(X,Y) = \frac{Cou(X,Y)}{\sqrt {Var(X) Var(Y)}}$$
- -1 <= ρ <= 1, X와 Y가 독립이면 ρ = 0 이다

$$ r = \frac
{{\sum_{i=1}^n}(x_i - \overline x)(y_i - \overline y)}
{
\sqrt{{\sum_{i=1}^n}(x_i - \overline x)^2}
\sqrt{{\sum_{i=1}^n}(y_i - \overline y)^2}
} $$

```r
# mrcars 데이터를 이용한 상관분석의 예
# install.packages('Hmisc')

library(Hmisc)
data(mtcars)
head(mtcars)
```

##### 나. 스피어만 상관계수(Spearman's rank correlation coefficient)

- 피어슨의 상관계수는 두 변수 간의 선형관계의 크기를 측정하는 값으로 "비선형적인 상관관계"는 나타내지 못함
- 스피어만 상관계수는 두 변수 간의 비선형적인 관계도 나타낼 수 있는 값으로, 
    - 한 변수를 단초증가 함수로 변환하여 다른 변수를 나타낼 수 있는 정도를 나타냄
    - 스피어만의 상관계수는 두 변수를 모두 순위로 변환시킨 후, 두 순위의 사이의 피어슨 상관계수로 정의됨
- 수식은 다음과 같음.

$$ \theta = \frac {\sum(r_i - \overline r)(s_i - \overline s)} {\sqrt \sum(r_i - \overline r)^2 \sqrt \sum(s_i - \overline s)^2 } , -1 <= \theta <= 1$$

- r_i 는 {x_1, x_2, ..., x_n} 에서 x_i의 순위
- s_i 는 {y_1, y_2, ..., y_n} 에서 y_i의 순위 

```r
rcorr(as.matrix(mtcars), type = "spearman")
```

```r
# 다음은 국어, 수학, 영어, 과학 점수를 데이터 프레임 형식으로 만든 데이터셋트.
korean <- c(85,75,65,78,59, 60,90,100,99,91, 70)
math <- c(80,60,75,40,50, 64,70,78,90,98, 50)
english <- c(80,70,69,79,80, 95,98,97,67,80, 59)
science <- c(90,100,50,80,67, 89,60,79,89, 80,100)
test <- data.frame(korean, math, english, science)
test
```

```r
rcorr(as.matrix(test), type="spearman")
```

#### 2. 다차원 척도법

- 다차원척도법(Multidimensional Scaling, MDS) : 여러 대상 간의 거리가 주어져 있을 떄, 대상들을 동일한 상대적 거리를 가진 실수 공간의 점들로 배치시키는 방법
    - 주어진 거리는 추상적인 대상들간의 거리가 될 수 있고, 실수 공간에서의 거리가 될 수 있음
    - 대상들을 2차원이나 3차원 실수공간의 점으로 대응시킬 수 있다면 이 점들을 시각화 할 수 있음
    - 이는 관측치들 간의 전반적 관계에 대한 직관적 이해를 할 수 있게 도와줌
    - 주로 자료들의 상대적 관계를 이해하는 시각화 방법의 근간으로 주로 사용됨

```r
data(eurodist)
eurodist
```

```r
loc <- cmdscale(eurodist)
loc
```

```r
x <- loc[,1]
y <- loc[,2]
plot(x, y, type="n", main="eurodist")
text(x, y, rownames(loc), cex = 0.8)
abline(v=0, h=0)
```

- eurodist 는 각 도시 사이의 거리를 포함하는 행렬형태의 자료
- cmdscale 함수를 사용하여 각 도시의 상대적 위치를 도식화 할 수 있는 x,y좌표를 계산하고 그래프로 표현
- 각 개체에 대한 특저 변수들의 관측지는 없더라도 객체 간의 유사성에 대한 자료를 사용하여 산점도를 그릴수 있음

#### 3. 주성분 분석

- 주성분 분석(Pricipal Component Analysis, PCA) : 상관관계가 있는 고차원 자료를 자료의 변동을 최대한 보존하는 저차원 자료로 변환 시키는 방법
- 자료의 차원을 축약시키는데 주로 사용됨

```r
library(datasets)
data(USArrests)
```

```r
summary(USArrests)
```

```r
fit <- princomp(USArrests, cor= TRUE)
summary(fit)
```

```r
loadings(fit)
```

```r
plot(fit, type = "lines")
```

- 주성분분석은 R 함수 princomp를 이용하여 수행되었고, 그 결고는 fit이라는 이름으로 저장됨
- cor=TRUE 옵션은 주성분분석을 공분행렬이 아닌 상관계수 행렬을 사용하여 수행한다
- summary(fit) 의 결과는 4개의 주성분의 표준편차, 분산의 비율 등을 보여줌
    - 첫번째 주성분 하나가 전체 분산의 약 62%를 설명
    - 두번째와 세번째 주성분의 분산은 전체 분산 중에 각각 25%와 9%를 설명함
    - 세번째 주서분까지 포함시키면 전체 분산의 약 96%까지 설명할 수 있음
    - loadings(fit) 함수는 주성분들의 로딩 벡터를 보여줌
    

$$ Y_1 = -0.536 Mrder - 0.583 Assault - 0.278 UrbanPop - 0.543 Rape$$
$$ Y_2 =  0.418 Mrder + 0.188 Assault - 0.873 UrbanPop - 0.167 Rape$$

- plot(fit) 함수는 각 주성분의 분산의 크기를 그림으로 그려줌
    - 그림을 스크리 그림(Scree plot)이라고 함
    - 이를 통해 주성분의 분산의 감소가 급격하게 줄어들어 주성분의 개수를 늘릴 때 얻게 되는 정보의 양이 상대적으로 미미한 지점에서 주성분의 개수를 정하는 것이 하나의 방법
    - 그 외에도 주성분들이 설명하는 총 분산의 비율이 70~90% 사이가 되는 주성분의 개수를 선택하는 방법을 사용하기도 함

```r
head(fit$scores)
```

```r
biplot(fit)
```

- fit$scores 는 각 관측치를 주성분들로 표현한 값을 나타냄
- biplot는 관측치들을 첫번쨰와 두 번째 주성분의 좌표에 그린 그림
- 첫 번쨰 주성분(Comp.1)이 Assault, Murder, Rape 변수들에 대해 상대적으로 큰 가중치를 적용하여 계산된 것을 알 수 있음
- 두 번째 주성분은 UrbanPop와 상대적으로 평행하기 때문에 다른 변수들에 비해 UrbabPop의 영향을 크게 받아 구성된 것으로 보임
- 첫 번째 주성분의 값이 작을수록 세 가지 범죄 발생율이 큰 주이고 두 번째 주성분 값이 작을수록 도심인구 비율이 큰 주라고 해석할 수 있음
